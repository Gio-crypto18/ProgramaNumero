```python
class Computadora:
    def __init__(self,marca,chip):
        self.marca = marca 
        self.chip = chip


    def __str__ (self):
        return f"marca{self.marca} con procesador {self.chip}"
    

class Laptop(Computadora):
    def __init__(self,marca,chip,precioBase,pantalla ="Normal"):
        super().__init__(marca,chip)
        self.pantalla = pantalla
        self.precioBase = precioBase
        self.incremento = {
            "i3":self.precioBase * 0.00,
            "i5":self.precioBase * 0.10,
            "i7":self.precioBase * 0.15
        }
        self.precioFinal = 0


    def precio(self):
        self.precioFinal = self.precioBase + self.incremento[self.chip]
        if self.pantalla == "Tactil":
            self.precioFinal = self.precioFinal +self.precioBase * 0.05
        self.precioFinal *= 1.18
        return self.precioFinal


    def __str__(self):
        return(
            super().__str__()
            + f", pantalla {self.pantalla}, precio final {self.precioFinal}"
        )


class Desktop(Computadora):
    def __init__(self,marca,chip,precioBase,pantalla=None):
        super().__init__(marca,chip)
        self.pantalla = pantalla 
        self.precioBase = precioBase
        self.incremento = {
            "i3": self.precioBase * 0.00,
            "i5": self.precioBase * 0.08,
            "i7": self.precioBase * 0.12
        }
        self.precioFinal = 0


    def precio(self):
        self.precioFinal = self.precioBase + self.incremento[self.chip]
        if self.pantalla == "HD":
            self.precioFinal + self.precioFinal + 200
        elif self.pantalla == "4K":
            self.precioFinal + self.precioFinal + 300
        else :
            self.pantalla = "No incluye pantalla"
        self.precioFinal *= 1.18
        return self.precioFinal
    

    def __str__(self):
        return(
            super().__str__()
            + f", pantalla {self.pantalla}, precio final {self.precioFinal}"
        )
    

    def listaPorTipo(lista,tipo):
        print(f"\nListado por tipo {tipo}")
        for compu in lista :
            if type(compu).__name__ == tipo:
                print(f"{compu}")


    def obtenerMayor(lista,mayor):
        print("\nPrecio maximo")
        for compu in lista:
            if compu.precioFinal == mayor :
                print(f"{type(compu).__name__}{compu}")


listaPrecios = []
laptop1 = Laptop ("Dell","i5",100)
listaPrecios.append(laptop1.precio())
laptop2 = Laptop ("HP","i3",100,"Tactil")
listaPrecios.append(laptop2.precio())
laptop3 = Laptop ("Dell","i7",100,"Tactil")
listaPrecios.append(laptop3.precio())


desktop1 = Desktop("Lenovo","i7",600,"4k")
listaPrecios.append(desktop1.precio())
desktop2 = Desktop("Lenovo","i5",600,"HD")
listaPrecios.append(desktop2.precio())
desktop3 = Desktop("Lenovo","i3",600,)
listaPrecios.append(desktop3.precio())


listaCompu = [laptop1,laptop2,laptop3,desktop1,desktop2,desktop3]


listaCatalogo(listaCompu)
listaPorTipo(listaCompu,"Laptop")
listaPorTipo(listaCompu,"Desktop")


obtenerMayor(listaCompu,max(listaPrecios))
```
