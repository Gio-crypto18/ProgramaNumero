```python
class Auto:
    def __init__(self,marca,material,tipo,cilindrada):
        self.marca = marca
        self.material = material
        self.tipo = tipo
        self.cilindrada = cilindrada


    def __str__ (self):
        return f"marca{self.marca}, {self.material}, de tipo {self.tipo},y su cilindrada es de {self.cilindrada}"
    

class Sedan(Auto):
    def __init__(self,marca,material,tipo,cilindrada,precioBase):
        super().__init__(marca,material,tipo,cilindrada)
        self.precioBase = precioBase
        self.suspension = {
            "posterior" : self.precioBase + 1400,
            "delantera" : self.precioBase + 1500,
            "doble" : self.precioBase + 1800
        }


    def precio(self):
        self.precioFinal = self.precioBase + self.suspension[self.suspension]
        if self.suspension == "posterior":
            self.precioFinal = self.precioFinal + self.precioBase
        elif self.suspension == "delantera":
            self.precioFinal = self.precioFinal + self.precioBase
        else :
            self.suspension =="doble"
            self.precioFinal = self.precioFinal + self.precioBase
        return self.precioFinal
    

    def __str__ (self):
        return (
            super().__str__() 
            + f", suspension {self.suspension}, y su precio final es de: {self.precioFinal}"
        )
    

class Hatchback(Auto):
    def __init__(self,marca,material,tipo,cilindrada,precioBase):
        super().__init__(marca,material,tipo,cilindrada)
        self.precioBase = precioBase
        self.tiempo = {
            "2 seg":self.precioBase *0.20,
            "4 seg":self.precioBase *0.10
        }


    def precio(self):
        self.precioFinal = self.precioBase + self.tiempo[self.tiempo]
        if self.tiempo == "2 seg":
            self.precioFinal = self.precioFinal + self.precioBase
        else :
            self.tiempo =="4 seg"
            self.precioFinal = self.precioFinal + self.precioBase
        return self.precioFinal    
    

    def __str__ (self):
        return (
            super().__str__() 
            + f", tiempo en encender {self.tiempo}, y su precio final es de: {self.precioFinal}"
        )
    

class Convertible(Auto):
    def __init__(self,marca,material,tipo,cilindrada,precioBase):
        super().__init__(marca,material,tipo,cilindrada)
        self.precioBase = precioBase
        self.maletera = {
            "Maletera":self.precioBase + 5000
        }


    def precio(self):
        self.precioFinal = self.precioBase + self.maletera[self.maletera]
        if self.maletera == "Maletera":
            self.precioFinal = self.precioFinal + self.precioBase
        return self.precioFinal 
    

    def __str__ (self):
        return (
            super().__str__() 
            + f", el costo de su maletera es de: {self.maletera}, y su precio final es de: {self.precioFinal}"
        )
    

def listaResumen(lista):
    print("\nCatalogo")
    for auto in lista:
        print(f"{type(auto).__name__} {auto}")


def listaPorTipo(lista,tipo):
    print(f"\nListado por tipo {tipo}")
    for auto in lista:
        if type(auto).__name__ == tipo:
            print(f"{auto}")


def obtenerMenor(lista,menor):
    print("\nPrecio minimo")
    for auto in lista:
        if auto.precioFinal == menor:
            print(f"{type(auto).__name__}{auto}")



listaMonto = []   
auto1 = Sedan ("aluminio","automatica","16cm3",1500,30000)
listaMonto.append(auto1.precio())
auto2 = Hatchback ("hierro","mecanica","Tactil","20cm3","2seg",30000)
listaMonto.append(auto2.precio())
auto3 = Convertible ("aluminio","electrico","Tactil","15cm3","Maletera",30000)
listaMonto.append(auto3.precio())



listaAuto = [auto1,auto2,auto3]

listaResumen(listaAuto)
listaPorTipo(listaAuto,"Auto")

obtenerMenor(listaAuto,min(listaMonto))
```
