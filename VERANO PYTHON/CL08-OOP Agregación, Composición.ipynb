### ***Ejemplo1: Agregación***
- Se debe crear una biblioteca, que tendrá como atributos el nombre de la biblioteca y la ciudad en donde está ubicada
- La biblioteca debe tener una lista donde almacenará los libros
- Dispondrá de un método mostrarBiblioteca que mostrará el nombre de la biblioteca, la ciudad y los libros si los hubiera
- Si no hubiera libros mostrará el mensaje "Dona libros a la biblioteca"
- Asimismo tendrá un método para agregar libros, chequeando que lo enviado sea un objeto
- Crear la clase Libro, con atributos título y autor y un método str
- El programa principal debe crear la biblioteca
- Además, se debe crear dos libros y agregarlos a la biblioteca
- y luego mostrar la biblioteca y sus libros

```python
class Libro:
    def __init__(self,titulo,autor):
        self.titulo = titulo
        self.autor = autor

    def __str__ (self):
        return f"{self.titulo} por {self.autor}"


class Bibilioteca:
    def __init__(self,nombre,ciudad):
        self.nombre = nombre
        self.ciudad = ciudad
        self.listaLibro = []


    def MostrarBiblioteca(self):
        print(f"{self.nombre} de la ciudad de {self.ciudad}")
        if len(self.listaLibro):
             for libro in self.listaLibro:
                 print(libro)
        else:
            print("Dona libros a la biblioteca")


    def AgregarLibro(self,libro):
        if isinstance(libro,Libro):
            self.listaLibro.append(libro)
        else:
            print("No has puesto un libro")


#Principal
libro1 = Libro("Los tres mosqueteros","Alejandro Dumas")
libro2 = Libro("El libro Troll","Ruben Doblas ")

biblioteca1 = Bibilioteca("Biblioteca Nacional","Lima")
biblioteca1.MostrarBiblioteca()

biblioteca1.AgregarLibro(libro1)
biblioteca1.AgregarLibro(libro2)
biblioteca1.MostrarBiblioteca()

biblioteca1.AgregarLibro("Holi")
```

### ***Ejemplo 2: Composición***
- Se creará un automovil que contendrá un motor
- Los atributos del motor son el número de cilindros y la potencia expresada en hp
- Tendrá un método "trabaja" que imprimirá un rango de números desde 0 hasta el número de hp
- La clase auto tendrá como atributos el modelo y el precio, además de requerir los datos para la creación del motor
- Tendrá un método "encender" que invocará al método "trabaja" del auto

```python
class Motor:
    def __init__(self,cilindros,hp):
        self.cilindros = cilindros
        self.hp = hp

    def __str__(self):
        return f"{self.cilindros} cilindros tiene {self.hp} de potencia"

    def trabaja(self):
        for i in range(self.hp):
            print(i,end=", ")


class Auto:
    def __init__(self,modelo,precio,cilindros,hp):
        self.modelo = modelo 
        self.precio = precio 
        self.motor = Motor(cilindros,hp)

    def __str__(self):
        return f"El auto {self.modelo} con motor {self.motor} y cuesta {self.precio}"

    def Encender(self):
        print("Encender motor")
        self.motor.trabaja()


#Principal 2
auto1 = Auto("Ferrari",500000,8,300)
print(auto1)
auto1.Encender()
```

### ***Polimorfismo***
- Los objetos de una subclase heredan los atributos y métodos de la super clase y, por tanto, a priori tienen el mismo comportamiento que los objetos de la super clase.
- Pero la subclase puede definir nuevos atributos o métodos o reescribir los métodos de la super clase de manera que sus objetos presenten un comportamiento distinto. Esto último se conoce como ***sobrecarga***.
- De este modo, aunque un objeto de la sub clase y otro de la super clase pueden tener un mismo método, al invocar ese método sobre el objeto de la sub clase, el comportamiento puede ser distinto a cuando se invoca ese mismo método sobre el objeto de la super clase.
- Esto se conoce como ***polimorfismo***

### ***Ejemplo 3: Polimorfismo***

```python
# Polimorfismo y sobrecarga
class Persona:
    def __init__(self, nombre, apellido, añonac):
        self.nombre = nombre
        self.apellido = apellido
        self.añonac = añonac

    def ImpDatos(self):
        print(
            f" : {self.nombre}, Apellido: {self.apellido}, AñoNac: {self.añonac}"
        )


class Estudiante(Persona):
    def __init__(self, nombre, apellido, añonac, añoing):
        super().__init__(nombre, apellido, añonac)
        self.año_ingreso = añoing

    def ImpDatos(self):
        print(
            f"Nombre: {self.nombre}, Apellido: {self.apellido}, "
            f"Año Ingreso: {self.año_ingreso}"
        )


P1 = Persona("Juan", "Pérez", 1990)
Estudiante1 = Estudiante("Miguel", "Grau", 1980, 2020)
lista = []
lista.append(P1)
lista.append(Estudiante1)
for valor in lista:
    valor.ImpDatos()

```

### ***Ejemplo 5***

- Una empresa vende computadoras: laptop y desktop. Las marcas pueden ser Dell, Acer y Lenovo (para nuestro caso no influyen en el precio). 
- El precio base de una laptop son 1000 dólares. Dependiendo del procesador, que pueden ser i3, i5 e i7, el precio final puede variar. 
    - Si el procesador es i3 se mantiene el precio base
    - Con i5 el precio amumenta en 10%
    - Con i7 en 15% 
- Para el caso de las las desktop, el precio base es 600 dólares 
    - Si el procesador es i3 se mantiene el precio base
    - Con i5 el precio amumenta en 8%
    - Con i7 en 12% 
- Además las laptop pueden tener pantalla táctil o normal, en el caso de la táctil el precio se incrementa en 5%. 
- En el caso de las desktop la pantalla es opcional y en caso de requerirla se le suma el precio de la misma, el cual será 200 dólares si es HD y 300 dólares si es 4K. 
- El precio final se obtiene según la configuración y hay que considerar el IGV.
- Se pide
    - Implementar las clases necesarias, considerando Herencia, y registre al menos 3 laptos y 3 desktop
    - Listar todas las computadoras registradas con todas sus características y el precio final
    - Elaborar un método que retorne la cantidad de laptops o dektop registradas en la tienda
    - Mostrar en pantalla todas las características de la computadora con el precio final más alto

```python
class Computadora:
    def __init__(self,marca,chip):
        self.marca = marca
        self.chip = chip

    def __str__ (self):
        return f"marca{self.marca} con procesador {self.chip}"


class Laptop(Computadora):
    def __init__(self,marca,chip,precioBase,pantalla="Normal"):
        super().__init__(marca,chip)
        self.pantalla = pantalla
        self.precioBase = precioBase 
        self.incremento = {
            "i3": self.precioBase * 0.00,
            "i5": self.precioBase * 0.10,
            "i7": self.precioBase * 0.15 
        }
        self.precioFinal = 0 


    def precio(self):
        self.precioFinal = self.precioBase + self.incremento[self.chip]
        if self.pantalla == "Tactil":
            self.precioFinal = self.precioFinal + self.precioBase * 0.05
        self.precioFinal *= 1.18
        return self.precioFinal
    

    def __str__ (self):
        return (
            super().__str__() 
            + f", pantalla {self.pantalla}, precio final {self.precioFinal}"
        )
    

class Desktop(Computadora):
    def __init__(self,marca,chip,precioBase,pantalla=None):
        super().__init__(marca,chip)
        self.pantalla = pantalla
        self.precioBase = precioBase 
        self.incremento = {
            "i3": self.precioBase * 0.00,
            "i5": self.precioBase * 0.08,
            "i7": self.precioBase * 0.12
        }
        self.precioFinal = 0 


    def precio(self):
        self.precioFinal = self.precioBase + self.incremento[self.chip]
        if self.pantalla == "HD":
            self.precioFinal + self.precioFinal + 200
        elif self.pantalla == "4K":
            self.precioFinal + self.precioFinal + 300
        else:
            self.pantalla = "No incluye pantalla"
        self.precioFinal *= 1.18
        return self.precioFinal


    def __str__ (self):
        return (
            super().__str__() 
            + f", pantalla {self.pantalla},precio final {self.precioFinal}"
        )


def listaCatalogo(lista):
    print("\nCatalogo")
    for compu in lista:
        print(f"{type(compu).__name__} {compu}")


def listaPorTipo(lista,tipo):
    print(f"\nListado por tipo {tipo}")
    for compu in lista:
        if type(compu).__name__ == tipo:
            print(f"{compu}")


def obtenerMayor(lista,mayor):
    print("\nPrecio maximo")
    for compu in lista:
        if compu.precioFinal == mayor:
            print(f"{type(compu).__name__}{compu}")


listaPrecios = []   
laptop1 = Laptop ("Dell","i5",1000)
listaPrecios.append(laptop1.precio())
laptop2 = Laptop ("HP","i3",1000,"Tactil")
listaPrecios.append(laptop2.precio())
laptop3 = Laptop ("Acer","i7",1000,"Tactil")
listaPrecios.append(laptop3.precio())


desktop1 = Desktop("Lenovo","i7",600,"4k")
listaPrecios.append(desktop1.precio())
desktop2 = Desktop("Acer","i5",600,"HD")
listaPrecios.append(desktop2.precio())
desktop3 = Desktop("HP","i3",600)
listaPrecios.append(desktop3.precio())


listaCompu = [laptop1,laptop2,laptop3,desktop1,desktop2,desktop3]

listaCatalogo(listaCompu)
listaPorTipo(listaCompu,"Laptop")
listaPorTipo(listaCompu,"Desktop")

obtenerMayor(listaCompu,max(listaPrecios))
```
